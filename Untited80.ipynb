```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
Standard_data = scaler.fit_transform(X)
X = Standard_data
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# 6 Train Model

```python
!pip install lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
lz = LazyClassifier(verbose=1,ignore_warnings=True, custom_metric=None)
models,predictions = lz.fit(X_train, X_test, y_train, y_test)
models
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(penalty='none')
lg.fit(X_train,y_train)
y_pred=lg.predict(X_test)
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train,y_train)
y_pred_2=sv.predict(X_test)
print(classification_report(y_test,y_pred_2))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(max_depth=20,criterion='gini')
rfc.fit(X_train,y_train)
y_pred_3=rfc.predict(X_test)
print(classification_report(y_test,y_pred_3))
```
